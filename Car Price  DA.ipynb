Data Preprocessing

Use SelectKbest for numerical feature selection

```python
# Select features by 'mutual_info_regression'

# Import Librarey
from sklearn.feature_selection import SelectKBest,mutual_info_regression,f_regression,chi2

# Independent variables
Numerical_features = df[['Year','Engine_Size','Mileage','Doors','Owner_Count']].copy()
# Dependent variable
Target = df['Price']
# Iniate the k_best and set score _func,feature limit
K_best = SelectKBest(score_func=mutual_info_regression,k=5)
# Fit the features X,Y
selector = K_best.fit(Numerical_features,Target)
# Create Dataframe to show the score of each columns
scores_df = pd.DataFrame({'Columns':Numerical_features.columns,'Scores':K_best.scores_})
print(scores_df)
```

```python
# Select features by 'f_regression'

# Import Librarey
from sklearn.feature_selection import SelectKBest,mutual_info_regression,f_regression,chi2
# Independent variables
Numerical_features = df[['Year','Engine_Size','Mileage','Doors','Owner_Count']].copy()
# Dependent variable
Target = df['Price']
# Iniate the k_best and set score _func,feature limit
K_best = SelectKBest(score_func=f_regression,k=5)
# Fit the features X,Y
selector = K_best.fit(Numerical_features,Target)
# Create Dataframe to show the score of each columns
scores_df = pd.DataFrame({'Columns':Numerical_features.columns,'Scores':K_best.scores_})
print(scores_df)
```

One-Hot Encoding

```python
# Import Libraries
from sklearn.preprocessing import OneHotEncoder

# Category features
cat_features = df[['Fuel_Type','Transmission']].copy()
# Inialize encoder
ohe_encoder = OneHotEncoder(sparse_output=False)
# Fit and transform the features
encoder = ohe_encoder.fit_transform(cat_features)
# make encoded data as Dataframe
encoded_df = pd.DataFrame(encoder,columns=ohe_encoder.get_feature_names_out(cat_features.columns))
encoded_df.head(3)
```

Feature Scaling

```python
# Import standard scaler Liberiery
from sklearn.preprocessing import StandardScaler

sd_scaler = StandardScaler()

Numerical_features = df[['Year','Engine_Size','Mileage']]
scaler = sd_scaler.fit_transform(Numerical_features)
scaler_df = pd.DataFrame(scaler,columns=sd_scaler.get_feature_names_out(Numerical_features.columns))
print('Parameters od StandardScaler:   ',sd_scaler.get_params())
print(scaler_df.head())


# Og Data
OG_df = sd_scaler.inverse_transform(scaler_df)
OG_df = pd.DataFrame(OG_df,columns=sd_scaler.get_feature_names_out(Numerical_features.columns))
print(OG_df.head())

```

```python
# Target Value 'Y'

sd_scaler_num = StandardScaler()
# Scaling Value 
Target = df[['Price']]
scaler_target = sd_scaler_num.fit_transform(Target)
scaled_target = pd.DataFrame(scaler_target,columns=['Price'])
print(scaled_target.head())

# OG Target
OG_Target = sd_scaler_num.inverse_transform(scaled_target)
OG_Target = pd.DataFrame(OG_Target,columns=['Price'])
print(OG_Target.head())
```

Concate Preprocessed Data

```python
# Concat Preprocesed data
car_df = pd.concat([encoded_df,scaler_df],axis=1)
car_df = car_df[['Year','Engine_Size','Fuel_Type_Diesel','Fuel_Type_Electric','Fuel_Type_Hybrid','Fuel_Type_Petrol','Transmission_Automatic','Transmission_Manual','Transmission_Semi-Automatic','Mileage']]
car_df.head()
```

Linear Regression Model Building

```python
# Impot libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error

# set features & Target
X = car_df.copy()
y = scaled_target

# Split Train & Split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)

# Initialize the Linearregression model
lr_model = LinearRegression()
# Fit the model for Trainning
lr_model = lr_model.fit(x_train,y_train)
# Predict the model
y_predict = lr_model.predict(x_test)

# Model Evaluation 
print('Accuracy: ',r2_score(y_predict,y_test))
# print('Mean squared error: ',mean_squared_error(y_predict,y_test))
print('Root mean squared error: ',root_mean_squared_error(y_predict,y_test))

# Check underfitting Preddiction on Trainning data
print('Predict on train Data')
x_predict = lr_model.predict(x_train)
print('Accuracy: ',r2_score(x_predict,y_train))
# print('Mean squared error: ',mean_squared_error(y_predict,y_test))
print('Root mean squared error: ',root_mean_squared_error(x_predict,y_train))


print('LR - Parameters: ',lr_model.get_params())
```

Parameters implementation

```python
# Impot libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error

# set features & Target
X = car_df.copy()
y = scaled_target

# Split Train & Split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)

# Initialize the Linearregression model
lr_model = LinearRegression(positive=False,n_jobs=1,fit_intercept=True,copy_X=True)
# Fit the model for Trainning
lr_model = lr_model.fit(x_train,y_train)
# Predict the model
y_predict = lr_model.predict(x_test)

# Model Evaluation 
print('Accuracy: ',r2_score(y_predict,y_test))
# print('Mean squared error: ',mean_squared_error(y_predict,y_test))
print('Root mean squared error: ',root_mean_squared_error(y_predict,y_test))

# Check underfitting Preddiction on Trainning data
print('Predict on train Data')
x_predict = lr_model.predict(x_train)
print('Accuracy: ',r2_score(x_predict,y_train))
# print('Mean squared error: ',mean_squared_error(y_predict,y_test))
print('Root mean squared error: ',root_mean_squared_error(x_predict,y_train))


print('LR - Parameters: ',lr_model.get_params())
print('Coeffcient value: ',lr_model.coef_[0][0])
```

App Building

```python
# Import Neccsary Libraries
import pandas as pd
import numpy as np
import pickle
import gradio as gr

# Import Gold Price Prediction Model
predict_ml = pickle.load(open(r'D:\workpads\MDTE15\ML mode trainning\Car Price project\Car Price Predictio Model.pkl','rb'))
print(type(predict_ml))


# Make it as Function
def car_price_prediction(Year,Engine_Size,Fuel_Type,Transmission,Mileage):
    User_input = pd.DataFrame({'Year':[Year],
                           'Engine_Size':[Engine_Size],
                           'Fuel_Type':[Fuel_Type],
                           'Transmission':[Transmission],
                           'Mileage':[Mileage],
                           })
    # Scaling the Numerical Data        
    num_feature= User_input[['Year','Engine_Size','Mileage']]
    num_feature2 = sd_scaler.transform(num_feature)
    final_num = pd.DataFrame(num_feature2,columns=sd_scaler.get_feature_names_out(num_feature.columns))

    # Encode Categorical Data
    cat_feature1 = User_input[['Fuel_Type','Transmission']]
    cat_feature2 = ohe_encoder.transform(cat_feature1)
    final_cat = pd.DataFrame(cat_feature2,columns=ohe_encoder.get_feature_names_out(cat_feature1.columns))

    # Concat preprocessing data as a dataframe
    main_features = pd.concat([final_num,final_cat],axis=1)
    main_features = main_features[['Year','Engine_Size','Fuel_Type_Diesel','Fuel_Type_Electric','Fuel_Type_Hybrid','Fuel_Type_Petrol','Transmission_Automatic','Transmission_Manual','Transmission_Semi-Automatic','Mileage']]

    # testing on user input
    Predicted_value = predict_ml.predict(main_features)
    # Inverse Value to Original Value
    OG_predicted_value = sd_scaler_num.inverse_transform(Predicted_value)
    return OG_predicted_value[0][0].round(2)
    


# App Interface
demo = gr.Interface(
    fn=car_price_prediction,
    inputs=["number", "number","text","text","number"],
    outputs=["number"],
)

demo.launch(share=True)
```
